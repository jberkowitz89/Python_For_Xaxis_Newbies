# Python Final Project

For our final project, each of you (individually) will either come up with a new analysis to complete using Python, recreate one of your existing pieces of analytics (a wrap, one sheeter, or otherwise), or automate a process using Python.

Use only Jupyter Notebooks to create your analysis. 

Use Python standard libraries, Pandas, Matplotlib, Seaborn or any other Python library you want to perform your analysis.

Each analysis must include at least three different data visualizations. Each analysis must also include takeaways (provided within the Jupyter Notebook as markdown notes.)

If you choose to automate a process, please details the steps you took, provide a sample of your inputs and outputs. 

**Googling and researching is encouraged! Use google, stack overflow, or any other resource you might need to work on your project**

### Project milestones
- Project Details sent to team: 4/15
- Project Proposal due: 4/23
- Project check-in: 4/30
- Final Presentations: 5/7

You will be evaluated on the quality of the analysis/process automation, as well as the your commentary and how well your code is explained/documented.
