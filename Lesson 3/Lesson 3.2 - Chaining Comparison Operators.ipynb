# Chained Comparison Operators

An interesting featureof Python is the ability to chain multiple comparisons to perform a more complex statement. You can use these chained comparisons as shorthand for larger Boolean Expressions
<br>
In this lecture, we will learn how to chain comparison operators and we will also introduce two other important statements in Python: **and** and **or**. 
<br>
Let's look at a few examples of using chained comparison operators.

```python
1 < 2 < 3
```

The above statement checks if 1 is less than 2 **and** if 2 is less than 3. We could have also used an **and** statement in Python.

```python
1 < 2 and 2 < 3
```

The **and** is used to make sure two checks have to be true in order for the total statement to be true. Let's see another example:

```python
1 < 3 > 2
```

The above checks if 3 is larger than both of the other numbers, so you could use **and** to rewrite it.

```python
1 < 3 and 3 > 2
```

Python is checking both instances of the comparisons. We can also use **or** statements to write comparisons in Python. For example:

```python
1 == 2 or 2 < 3
```

Note how this statement was true; With the **or** operator, only one of the statements need to be true. Another example:

```python
1 == 1 or 100 == 1
```

Now, we should have an understanding of all the different comparison operators, and using **and** and **or** statements.
