### Translate the following into Python "if/else" statements


```python
year = 2020

#If the year is divisible by 4, print "Could be a leap year". Otherwise, print "Definitely not a leap year".

#Your code here
if year % 4 == 0:
    print('Could be a leap year')
else:
    print('Definitely not a leap year')
```

```python
ticket = [1, 14, 34, 45, 59]
lottery = [1, 14, 33, 45, 59]

#If ticket is equal to lottery, print "You won!". Else, print "Better luck next time"

#Your code here

if ticket == lottery:
    print('You won!')
else:
    print('Better luck next time')
```

### Translate the following into a Python "if/elif/else" statement

Take the current temperature. <br>
If the temperature is greater or equal to 75 degrees, print "It is hot, be sure to drink liquids". <br>
If the temperature is less than 75 degrees, but above or equal to 50 degrees, print "It is comfortable, but please pack a light jacket". <br>
If the temperature is less than 50 degrees but above or equal to 30, print "It is getting colder, please dress warmer" <br>
If the temperature is less than 30 degrees, print "It is freezing! Please stay warm!"



```python
#Your code here
temp = 60

if temp >= 75:
    print('It is hot, be sure to drink liquids')
elif temp < 75 and temp >= 50:
    print('It is comfortable, but please pack a light jacket')
elif temperature < 50 and temp >= 30:
    print('Itis getting colder, please dress warmly')
else:
    print('It is freezing! Please stay warm!')
```
