**Use <code>for</code>, <code>.split()</code>, and <code>if</code> to create a statement taht will print out words that start with 's':**

```python
st = 'Print only the words that start with s in this sentence'
```

```python
#Your Code here
for word in st.split():
    if word[0] == 's':
        print(word)
```

**Question 5: Write a program that prints the integers from 1 to 100. But for multiples of three print "Fizz" instead of the number, and for the multiples of five print "Buzz". For numbers which are multiples of both three and five print "FizzBuzz".**

```python
# Your Code Here
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 5 == 0:
        print("Buzz")
    elif i % 3 == 0:
        print("Fizz")
    else:
        print(i)
```
