# Dictionaries

This section will serve as a brief introduction to dictionaries and consist of:

    1.) Constructing a Dictionary
    2.) Accessing objects from a dictionary
    3.) Nesting Dictionaries
    4.) Basic Dictionary Methods

A Python dictionary consists of a key and then an associated value. The value can be almost any Python object.

## Constructing a Dictionary

```python
#Making a dictionary with {}

my_dict = {'key1':'value1', 'key2':'value2'}
print(my_dict)
```

```python
#Call values by their key
my_dict['key2']
```

Dictionaries are very flexible in terms of the data types that they can hold.

```python
my_dict = {"key1":123, "key2":[12,23,33], "key3": ["item0", "item1", "item2"]}
```

```python
#Call items from the dictionary
my_dict['key3']
```

```python
#Can call an index on that value
my_dict["key3"][0]
```

```python
#Can call methods on values
my_dict["key3"][0].upper()
```

We can updated and affect values of a key. For example:

```python
my_dict["key1"]
```

```python
#Subtract 123 from that value
my_dict["key1"] = my_dict["key1"] - 123
```

```python
my_dict["key1"]
```

Python also has a method of doing self subtraction, addition, multiplication or subtraction. We can use -= or +=!

```python
#Set the object equal to itself plus 123
my_dict["key1"] += 123
my_dict["key1"]
```

We can also create keys by assignment. For instance, we can start off with an empty dictionary and continually add to it.

```python
#Create a new Dictionary
d = {}
```

```python
#Create a new key through assignment
d['animal'] = "Dog"
```

```python
d
```

```python
d["answer"] = 42
```

```python
d
```

## Nesting with Dictionaries

We can nest dictionaries inside other dictionaries (just like lists).

```python
d = {"key":{"nestkey":{"subnestkey":"value"}}}
```

How would we get to that value?

```python
d["key"]["nestkey"]["subnestkey"]
```

## Dictionary Methods
There are a few important methods to know for dictionaries.

```python
#Create a dictionary
d = {"key1":1, "key2":2, "key3":3}
```

```python
#Method to return a list of all keys
d.keys()
```

```python
#Method to grab all values
d.values()
```

```python
#Method to return tuples of all items
d.items()
```
