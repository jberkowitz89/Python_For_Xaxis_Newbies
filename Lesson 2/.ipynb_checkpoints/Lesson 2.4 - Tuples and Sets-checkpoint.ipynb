# Tuples

In Python tuples are very similar to lists, however, unlike lists they are *immutable* meaning they can not be changed. You would use tuples to present things that shouldn't be changed. 

In this section, we will get a brief overview of the following:

    1.) Constructing Tuples
    2.) Basic Tuple Methods
    3.) Immutability
    4.) When to Use Tuples

You'll have an intuition of how to use tuples based on what you've learned about lists. We can treat them very similarly with the major distinction being that tuples are immutable.

### Constructing Tuples

The construction of a tuples use ( ) with elements separated by commas. For example:

```python
#Creating a tuple
t = (1,2,3)
```

```python
#Check len just like a list
len(t)
```

```python
#Can also mix object types within a tuple

t = ('one', 2)

t
```

```python
#Can use indexing and slicing just like lists

t[0]
```

### Basic Tuple Methods

```python
#use .index to enter a value and return the index

t.index('one')
```

```python
#Use .count to count the number of timesa value appears
t.count('one')
```

### Immutability
TUPLES CANNOT BE CHANGED - AKA THEY ARE IMMUTABLE!

```python
t[0] = 'CHANGE'
```

Because of this, tuples can't grow. Once a tuple is made, we can't add to it.

### When to use Tuples

Tuples are not used as often as lists, but are used when immutability is necessary. If you are passing around data or an object within your program and want to make sure that it doesn't get changed, then tuples are your solution. 

# Sets

Sets are an unordered collection of **unique** elements. We can construct them using the set() functoin.

```python
x = set()
```

```python
#We can add to sets with the add() method

x.add(1)
```

```python
x
```

Though there are curly brackets, x is not a dictionary! 
<br>
Since we know that sets are made up of just unique elements, let's see what happens when we try to add a duplicate element to a set.

```python
x.add(2)
```

```python
x
```

```python
x.add(1)
```

```python
x
```

Sets can be useful for - you guessed it - finding the unique elements in a list.

```python
#Creating a list with repeats

list1 = [1,1,2,2,3,4,5,6,1,2]
```

```python
list1
```

```python
set(list1)
```
