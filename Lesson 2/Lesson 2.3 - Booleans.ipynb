# Booleans

Booleans are an object in Python that denote a condition - denoted by True or False. It alsohas a placeholder object called None. We'll work with booleans more once we get to control flow, but for now, let's go through a few examples.

```python
#Set object to be a boolean

a = True
```

```python
a
```

```python
type(a)
```

We can also use comparison operators to create booleans. We'll learn more about comparison operators later on.


```python
1 > 2
```

```python
b = 1 > 2
```

```python
b
```

We can also use None as a placeholder for a variable that we want to define, but don't want to assign value to yet.

```python
#None placeholder
b = None
```

```python
print(b)
```
