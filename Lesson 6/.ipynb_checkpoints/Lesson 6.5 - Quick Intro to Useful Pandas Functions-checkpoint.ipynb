# Quick Intro to Useful Pandas Functions
Pandas has a ton of useful functions, but quickly I wanted to show you all some of the most useful ones that I use most frequently.

```python
import pandas as pd
df = pd.read_csv("ctr_by_state.csv")
```

### head()
The head() function shows you the first n rows of a dataframe or series. The default number of rows is 5, but you can specify if you want to see more rows.

```python
df.head()
```

```python
df.head(10)
```

### tail()
The tail() function does the opposite - it shows you the last rows of a dataframe.

```python
df.tail()
```

```python
df.tail(10)
```

### describe()
The describe function gives you a really quick, easy one-line way to get some descriptive statistics about your data.

```python
pd.set_option('display.float_format', lambda x: '%.5f' % x)
df.describe()
```

### value_counts()
The value_counts() function give the number of occurrences of all values for a single series.

```python
df["State"].value_counts()
```

### shape
The shape attribute gives us the shape of our dataframe - how many rows and columns we have in our data

```python
df.shape
```

### info
The .info() attribute lists all of the data types and the number of non-missing values in each column. It gives a good snapshot into our data, and lets us know which columns might be problematic when it comes to missing values or incorrectly encoded data types.

```python
df.info()
```
