```python
import pandas as pd
```

```python
df=pd.read_excel("Creative_Wrap_20200506_231949_667469370_2897403254.xlsx")
```

```python
df.head()
```

```python
df.columns= ['Day','Site ID','Site','Advertiser ID','Advertiser','Currency','Creative Name','Creative Key','Placement','Placement ID','Impressions','Clicks','TrueViews','TrueView Rate','Video Fully Played','Video Completion 25%','Video Completion 50%','Video Completion 75%']
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.drop(df.columns[[1,2,3,4,5,7,9,13]],axis=1,inplace=True)
```

```python
df.head()
#An ideal location to add IO rate and rate type
```

```python
df.loc[df['Placement'].str.contains('Bumper'), 'Package'] = 'Bumper'
```

```python
df.loc[df['Placement'].str.contains('Trueview'), 'Package'] = 'TrueView Classic'
```

```python
import numpy as np
pivot=pd.pivot_table(data=df, index='Package', aggfunc=np.sum)
pivot
#other possible pivots: creative and rate type
```

```python
pivot['VTR']= (pivot["TrueViews"]/pivot["Impressions"]*100).round(2)
```

```python
pivot['VCR']= (pivot["Video Fully Played"]/pivot["Impressions"]*100).round(2)
```

```python
pivot.loc[['Bumper'], 'Type'] = 'BMP'
```

```python
pivot.loc[['TrueView Classic'], 'Type'] = 'TRV'
```

```python
pivot
```

```python
pivot_day=pd.pivot_table(data=df, index=['Package','Day'], aggfunc=np.sum)
```

```python
pivot_day.head()
```

```python
pivot_day['VTR']= (pivot_day["TrueViews"]/pivot_day["Impressions"]*100).round(2)
```

```python
pivot_day['VCR']= (pivot_day["Video Fully Played"]/pivot_day["Impressions"]*100).round(2)
```

```python
pivot_day.head()
```

```python
pivot_day.loc[['Bumper'], 'Benchmark'] = 90.0
```

```python
pivot_day.loc[['TrueView Classic'], 'Benchmark'] = 30.0
```

```python
pivot_day.loc[['Bumper'], 'VTR/VCR'] = (pivot_day["Video Fully Played"]/pivot_day["Impressions"]*100).round(2)
```

```python
pivot_day.loc[['TrueView Classic'], 'VTR/VCR'] = (pivot_day["TrueViews"]/pivot_day["Impressions"]*100).round(2)
```

```python
pivot_day['Over Benchmark']=((pivot_day["VTR/VCR"]-pivot_day['Benchmark'])/pivot_day["Benchmark"]*100).round(2)
```

```python
pivot_day.loc[['Bumper'], 'Type'] = 'BMP'
```

```python
pivot_day.loc[['TrueView Classic'], 'Type'] = 'TRV'
```

```python
pivotable_day=pivot_day.reset_index()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
ax = pivot.plot.bar(x='Type', y=['VCR','VTR'], rot=0)
```

```python
zy=pivotable_day.loc[pivotable_day["Package"]=="TrueView Classic","Over Benchmark"]
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(xy, y, label="VTR", color="#8B008B")    
ax.plot(xy, zy, label="%Over", color="#FF8C00")
plt.xticks(rotation=45)
ax.legend()
```
